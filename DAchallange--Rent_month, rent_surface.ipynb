### Average monthly rent (€ / month) and average rent per surface (€ / m2 per month) of the city of Barcelona

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import calendar
sns.set()
```

```python
df2014 = pd.read_csv('./data/2014.csv')
df2015 = pd.read_csv('./data/2015.csv')
df2016 = pd.read_csv('./data/2016.csv')
df2017 = pd.read_csv('./data/2017.csv')
df2018 = pd.read_csv('./data/2018.csv')
df2019 = pd.read_csv('./data/2019.csv')
print('shape of 2014 data:', df2014.shape)
print('shape of 2015 data:', df2015.shape)
print('shape of 2016 data:', df2016.shape)
print('shape of 2017 data:', df2017.shape)
print('shape of 2018 data:', df2018.shape)
print('shape of 2019 data:', df2019.shape)
```

we have 6 years data for rent prices. each dataset has 584 rows and 8 columns. we will be concating all the datasets together

```python
df = pd.concat([df2014,df2015,df2016,df2017,df2018,df2019])
```

```python
df.shape
```

so in complete dataset we have 3504 rows with 8 columns. in our dataset we have 2 types of prices we will be sepreating our data in those 2 categories

```python
df.Lloguer_mitja.value_counts()
```

so we have 1st category where rent is average monthly , while 2nd category is average rent per area.

```python
df1 = df.loc[df['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)'] # average rent per month
df2 = df.loc[df['Lloguer_mitja']=='Lloguer mitjà per superfície (Euros/m2 mes)'] # average rent per area
   
```

```python
print(df1.shape)
print(df2.shape)
```

```python
print(df1.Preu.isna().sum())
print(df2.Preu.isna().sum())
```

we have missing values in our rent column we will be filling out these missing values with mean 

```python
# filling missing values with mean
df1 = df1.fillna(df1.Preu.mean())
df2 = df2.fillna(df2.Preu.mean())
```

we need to see the maximum rent value for each year for each district and similarly the minimum value

```python
# function for maximum value by year
def maximum_rent(df,year):
    a = df.loc[df.Any == year].groupby('Nom_Districte').max()
    x = a.index
    y = a['Preu']
    plt.barh(x,y,color = color)
    plt.title(f'Barcelona maximum rent distibution for each district for year{year}')
    plt.xlabel('rentvalues')
    plt.ylabel('district_names')
    return plt.show()
```

```python
# function for minimum value per year
def minimum_rent(df,year):
    a = df.loc[df.Any == year].groupby('Nom_Districte').min()
    x = a.index
    y = a['Preu']
    plt.barh(x,y,color = color)
    plt.title(f'Barcelona minimum rent distibution for each district for year{year}')
    plt.xlabel('rentvalues')
    plt.ylabel('district_names')
    return plt.show()
```

```python
# function for average rent value per year
def average_rent(df,year):
    a = df.loc[df.Any == year].groupby('Nom_Districte').mean()
    x = a.index
    y = a['Preu']
    plt.barh(x,y,color = color)
    plt.title(f'Barcelona average rent distibution for each district for year{year}')
    plt.xlabel('rentvalues')
    plt.ylabel('district_names')
    return plt.show()
```

```python
maximum_rent_per_month_2014 = maximum_rent(df1,2014)
maximum_rent_per_area_2014 = maximum_rent(df2,2014)
minimum_rent_per_month_2014 = minimum_rent(df1,2014)
minimum_rent_per_area_2014 = minimum_rent(df2,2014)
average_rent_per_month_2014 = average_rent(df1,2014)
average_rent_per_area_2014 = average_rent(df2,2014)
```

### models

```python
data = pd.concat([df1,df2])
```

```python
features = ['Trimestre','Codi_Districte','Codi_Barri']
target = data['Preu']
```

```python
## train test split
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(data[features],target,test_size=0.3,random_state=0)
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics

lr = LinearRegression()
lr.fit(X_train, y_train)

fitted_values = pd.DataFrame({
    'prediction': lr.predict(X_train)
}, index=X_train.index)

mse_in_sample = mean_squared_error(y_train, fitted_values)
print(f'Mean squared error in-sample: {mse_in_sample}')
```

```python
predictions_baseline = pd.DataFrame({
    'prediction': lr.predict(X_test)
}, index=X_test.index)
#mse_SalePrice = mean_squared_error(y_test, predictions_baseline)
losses1 = {'baseline': {}}
losses1['baseline']['MAE'] = metrics.mean_absolute_error(y_test, predictions_baseline)
losses1['baseline']['MSE'] = metrics.mean_squared_error(y_test, predictions_baseline)
losses1['baseline']['RMSE'] = np.sqrt(metrics.mean_squared_error(y_test, predictions_baseline))
losses1['baseline']['Accuracy'] = lr.score(X_test,y_test)
```
